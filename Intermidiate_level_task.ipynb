```python
# ---------------------------------------------------------
# 1. Imports and configuration
# ---------------------------------------------------------
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8-whitegrid")
sns.set_context("talk")

# ---------------------------------------------------------
# 2. Load data
#    Assumes delhiaqi-1.csv has columns:
#    ['date','co','no','no2','o3','so2','pm2_5','pm10','nh3']
# ---------------------------------------------------------
path = "/content/drive/MyDrive/Shadowfox/delhiaqi.csv"
df = pd.read_csv(path)

# Parse datetime and sort
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values("date")

# Create time-based features
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day
df["hour"] = df["date"].dt.hour

# List of pollutants
pollutants = ["pm2_5", "pm10", "no2", "so2", "co", "o3", "nh3", "no"]

# ---------------------------------------------------------
# 3. Descriptive statistics
# ---------------------------------------------------------
overall_stats = df[pollutants].describe().T
print("Overall descriptive statistics:")
print(overall_stats)

# ---------------------------------------------------------
# 4. Correlation matrix
# ---------------------------------------------------------
corr = df[pollutants].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="RdBu_r", vmin=-1, vmax=1, fmt=".2f")
plt.title("Correlation Matrix of Key Pollutants (Delhi, Jan 2023)")
plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 5. Daily averages (time series)
# ---------------------------------------------------------
daily = (
    df.set_index("date")[pollutants]
      .resample("D")
      .mean()
      .reset_index()
)

print("First few daily averages:")
print(daily.head())

# Plot daily time series for main pollutants
plt.figure(figsize=(15, 7))
for col in ["pm2_5", "pm10", "no2", "so2", "o3"]:
    plt.plot(daily["date"], daily[col], marker="o", label=col)

plt.title("Delhi Daily Average Pollutant Levels (Jan 2023)")
plt.xlabel("Date")
plt.ylabel("Concentration (µg/m³ or ppb equivalent)")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 6. Hour-of-day patterns (diurnal cycle)
# ---------------------------------------------------------
hourly = df.groupby("hour")[pollutants].mean().reset_index()
print("First few hourly means:")
print(hourly.head())

plt.figure(figsize=(14, 6))
for col in ["pm2_5", "pm10", "no2", "co"]:
    plt.plot(hourly["hour"], hourly[col], marker="o", label=col)

plt.title("Average Diurnal Pattern of Key Pollutants (Jan 2023)")
plt.xlabel("Hour of Day")
plt.ylabel("Mean Concentration")
plt.legend()
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# 7. Identify worst day by PM2.5 and PM10
# ---------------------------------------------------------
daily_pm = (
    df.set_index("date")[["pm2_5", "pm10"]]
      .resample("D")
      .mean()
)

worst_pm25_day = daily_pm["pm2_5"].idxmax()
worst_pm10_day = daily_pm["pm10"].idxmax()

print("Worst PM2.5 day:", worst_pm25_day, "value:", daily_pm.loc[worst_pm25_day, "pm2_5"])
print("Worst PM10 day:", worst_pm10_day, "value:", daily_pm.loc[worst_pm10_day, "pm10"])

# ---------------------------------------------------------
# 8. Optional: Boxplots to see distribution by pollutant
# ---------------------------------------------------------
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[pollutants])
plt.title("Distribution of Pollutant Concentrations (Hourly, Jan 2023)")
plt.ylabel("Concentration")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
