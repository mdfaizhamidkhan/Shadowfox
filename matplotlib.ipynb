#The subplot() Function

The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

#Matplotlib Scatter:
we can use the scatter() function to draw a scatter plot.

**Defination** -
A scatter plot displays the relationship between two numerical variables,
with each data point represented as a dot. It is essential for observing correlation,
clusters, and outliers.

````python
#Basic Approach:
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
#plt.plot(x,y)
plt.show()
````

````python
#ColorMap
import matplotlib.pyplot as plt
import numpy as np
from numpy import random

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.random.rand(len(x)) # Generate colors with the same number of elements as x and y
plt.scatter(x, y, c=colors, cmap='viridis')
plt.show()
````

````python
#Use Case: Comparing student study hours vs. their test scores, examining the
#relationship between a car's weight and its MPG, identifying clusters in data.
np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)
# Create the plot
plt.figure(figsize=(8, 4))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.6, cmap='viridis')
# Add labels and title
plt.title("Scatter Plot")
plt.xlabel("X Value")
plt.ylabel("Y Value")
# Add a color bar
plt.colorbar().set_label('Color Intensity')
````

#Matplotlib Bars
 we can use the bar() function to draw bar graphs.

````python
#example
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
````

#Histogram
(This is a type of frequency distribution )

It helps to  visualizes the distribution of a single numerical dataset. The
data is grouped into a series of intervals (bins), and the chart shows the frequency
(count) of data points falling into each bin.


````python
#Understanding the distribution of exam scores, visualizing the age
#distribution of a population, checking if data is normally distributed.
data = np.random.randn(100)
# Create the plot
plt.figure(figsize=(4, 4))
plt.hist(data, bins=30, edgecolor='red', alpha=0.7)
# Add labels and title
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
````

#Pie Charts:
 It is a circular(360 degree) graph divided into slices to illustrate numerical
proportion. While visually common, they are often criticized because it can be difficult
for humans to compare angles. A bar chart is often a better alternative.

````python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
````

````python
#: Showing market share of different companies, breaking down a budget by
#expense category (for a small number of categories).
sizes = [15, 30, 45, 10]
labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
explode = (0, 0.1, 0, 0) # "explode" the 2nd slice
# Create the plot
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, explode=explode, autopct='%1.1f%%',
shadow=True, startangle=140)
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.title("Simple Pie Chart")
````

#3. Comparison: Strengths & Weaknesses
**Strengths**
  * **Power & Customization**:  we have finegrained control over every single element of our plot. If you can imagine it, we
can build it, which is why it's the standard for publication-quality figures.
  * **Flexibility**: The Object-Oriented API is incredibly powerful. It allows us to build
complex, multi-plot figures and embed plots into GUI applications (like Tkinter, Qt,
or wxPython).
* **Vast Ecosystem & Community**: As the original Python plotter, it has a massive
community. This means we can find documentation, examples, and Stack Overflow
answers for almost any problem. It is also the backend for many other libraries (like
Seaborn), so understanding Matplotlib helps us to use those tools better.
* **Stability & Reliability**: It is mature, well-tested, and a reliable workhorse for
scientific computing.
